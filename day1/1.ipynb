```python
##装饰器
#核心作用：不修改原函数内部代码，动态的给函数新增功能
```

```python
def double_effect(func):
    def wrapper(*args, **kwargs):
        print("魔法特效：双倍伤害")
        res=func(*args, **kwargs)
        return res*2
    return wrapper

@double_effect
def attack_damage(base):
    return base

print(f"{attack_damage(100)}")
```

```python
import time
def log_record(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        res = func(*args, **kwargs)
        cost_time = time.time() - start_time
        print(f"调用{func.__name__},耗时{cost_time:.2f}s")
        return res
    return wrapper
@log_record
def calc_damage(base):
    return base*2

calc_damage(20)
```

```python
def log_required(func):
    def wrapper(user,*args, **kwargs):
        if not user:
            print("<请先登录>")
            return
        return func(user,*args, **kwargs)
    return wrapper

@log_required
def delete_item(user,item_id):
    print(f"删除道具{item_id}")

delete_item(True,12131241)
delete_item(False,12131241)
```

```python
cache={}
def cache_func(func):
    def wrapper(*args):
        if args in cache:
            print("命中缓存，直接读取")
            return cache[args]
        res=func(*args)
        cache[args] = res
        return res
    return wrapper

@cache_func
def big_calc(a,b):
    print("复杂计算中...")
    return a**b

big_calc(3,2)
print(cache)

big_calc(3,2)
```
