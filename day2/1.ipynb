```python
from itertools import count


class Car:
    def __init__(self, color, brand):
        self.color = color  #把传进来的color存到对象身上
        self.brand = brand  #把传进来的brand存到对象身上
        self.speed = 0

    def run(self, how_fast):
        self.speed = how_fast
        print(f"{self.color}的{self.brand}正在以{self.speed} km/h行驶")

    def stop(self):
        self.speed = 0
        print(f"{self.color}的{self.brand}停下来了")

my_car = Car("红色", "byd")
my_car.run(60)
my_car.stop()
```

注意：方法的第一个参数永远是self，但调用时不需要自己传，python会自动把对象本身塞进去

| 分类维度 | 类型 | 标识特征 | 核心特点 |
| :--- | :--- | :--- | :--- |
| 访问权限 | 公有方法 | 无下划线 | 外部任意调用，对外接口 |
| 访问权限 | 保护方法 | `_method` | 仅类 / 子类内部使用，软性私有 |
| 访问权限 | 私有方法 | `__method` | 外部无法直接访问，强隐藏 |
| 参数标识 | 实例方法 | 首参 `self` | 操作实例独有数据（90% 场景） |
| 参数标识 | 类方法 | `@classmethod`，首参 `cls` | 操作类共享属性 |
| 参数标识 | 静态方法 | `@staticmethod`，无默认参数 | 纯工具计算，不碰任何属性 |
| 内置魔术 | 魔术方法 | `__func__` 前后双下划线 | 创建 / 打印对象时自动触发 |
| 业务用途 | 抽象模板方法 | 父类抛异常 | 强制子类重写业务逻辑 |


1. 按权限分类
1.1 公有方法（无下划线）
定义：方法名不加_，外部代码可只有调用，是对外提供的功能接口
2. 保护方法(单下划线_xxx)
约定：仅限本类、子类内部使用，外部不建议调用，语法不强制拦截
3. 私有方法(双下划线__xxx)
外部、子类无法访问，完全用于内部底层逻辑

```python
class Test:
    def __calc(self):
        return 1000

t=Test()
t.__calc()
```

2. 按照第一个参数区分
2.1 实例方法： 最常见 特征：第一个固定参数self，只能用实例调用
2.2 类方法@classmethod 装饰器，第一个参数cls，代表整个类，只能操作类属性，不能访问实例独有的数据

```python
class PointAllocator:
    total_alloc_times = 0

    @classmethod
    def print_total(cls):
        print(f"所有点被总分配次数: {cls.total_alloc_times}")
```

```python
class student:
    teacher="丘老师"
    def __init__(self, name, age):
        self.name = name
        self.year = age

    def introduce(self):
        print(f"我叫{self.name}，今年{self.year}，我的老师叫{self.teacher}")

    @classmethod
    def change_name(cls,newteacher):
        cls.teacher = newteacher
        print(f"老师名统一为{cls.teacher}")

    @staticmethod
    def showtip():
        print("听讲！")

student.showtip()

lihua=student("李华",21)
lihua.introduce()

lihua.change_name("秦老师")

xiaoming=student("小明",21)
xiaoming.introduce()
```

```python
# 这是一个生成器函数
def card_dealer():
    print("🎲 准备发牌...")
    yield "红桃A"      # 暂停，交出第一张牌
    yield "黑桃K"      # 继续执行，暂停，交出第二张牌
    yield "方块Q"      # 继续执行，暂停，交出第三张牌
    print("🎲 牌发完了！")

# 调用生成器函数，得到一个生成器对象
my_cards = card_dealer()

# 我们手动去“拿”牌
print(next(my_cards))  # 输出: 红桃A (此时函数暂停在这里)
print("我拿到牌了，等一下再拿...")
print(next(my_cards))  # 输出: 黑桃K (函数从刚才暂停的地方继续)
print(next(my_cards))  # 输出: 方块Q
print(next(my_cards))  # 报错：StopIteration (牌发完了)
```

```python
# 准备一个普通的列表
my_list = ['苹果', '香蕉', '橘子']

# 第一步：把它变成一个迭代器（放进分发机里）
my_iter = iter(my_list)

# 第二步：手动去“拿”数据
print(next(my_iter))  # 输出: 苹果
print(next(my_iter))  # 输出: 香蕉
print(next(my_iter))  # 输出: 橘子

# 第三步：如果已经拿完了，再按一次按钮会怎样？
print(next(my_iter))  # 报错：StopIteration (停止迭代)
```

```python
import time

class AIModel:
    def __init__(self, name, model_type):
        self.name = name
        self.model_type = model_type

    def predict(self, input_data):
        print(f"{self.name}模型收到输入：{input_data}，但具体推理逻辑由子类实现")
        return "父类默认结果"

class TextModel(AIModel):
    # 重写 predict 方法
    def predict(self, input_data):
        time.sleep(1)
        print(f"文本模型{self.name}正在生成文本...")
        return f"生成的文本结果: {input_data}"


class ImageModel(AIModel):
    def predict(self, input_data):
        time.sleep(2)
        print(f"图像模型{self.name}正在识别图像...")
        return f"识别结果: {input_data}"


# 5. 创建实例并调用方法测试
if __name__ == "__main__":
    text_model = TextModel("文心一言", "NLP")
    text_result = text_model.predict("今天天气真好")
    print(text_result)
    image_model = ImageModel("通义万相", "CV")
    image_result = image_model.predict("一张猫的图片")
    print(image_result)
```

```python
import threading

count=0
lock=threading.Lock()

def add():
    global count
    for _ in range(100):
        lock.acquire()
        count+=1
        lock.release()

ts=[threading.Thread(target=add) for _ in range(5)]
for t in ts:
    t.start()

for t in ts:
    t.join()

print(count)
```

```python
from datetime import datetime
import time
import threading

records=[]
lock=threading.Lock()

def simulate_inference(model_name, seconds):
    start = time.time()

    print(f"{model_name}开始推理")
    time.sleep(seconds)

    end = time.time()
    cost=end-start
    print(f"推理结束")
    record = {
        "model": model_name,
        "start": datetime.fromtimestamp(start).strftime("%Y-%m-%d %H:%M:%S"),
        "end": datetime.fromtimestamp(end).strftime("%Y-%m-%d %H:%M:%S"),
        "cost": cost
    }

    with lock:
        records.append(record)

print("——————串行————————")
start1 = datetime.now()

simulate_inference("deepseek1",1)
simulate_inference("deepseek2",1)
simulate_inference("deepseek3",1)
simulate_inference("deepseek4",1)

end1=datetime.now()
print(end1-start1)
print("                        ")


print("-------并行-----------")
start2 = datetime.now()

threads=[]
for i in range(1,5):
    t=threading.Thread(target=simulate_inference,args=(f"qwen{i}",1))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

end2=datetime.now()
print(end2-start2)
print("                                 ")
for i in records:
    print(i)
```

```python
import time
import threading

records=[]
lock=threading.Lock()

class AIModel:
    def __init__(self, name, model_type):
        self.name = name
        self.model_type = model_type

    def predict(self, input_data):
        print(f"{self.name}模型收到输入：{input_data}，但具体推理逻辑由子类实现")
        return "父类默认结果"

class TextModel(AIModel):
    def predict(self, input_data):
        time.sleep(1)
        print(f"文本模型{self.name}正在生成文本...")
        return f"生成的文本结果: {input_data}"


class ImageModel(AIModel):
    def predict(self, input_data):
        time.sleep(2)
        print(f"图像模型{self.name}正在识别图像...")
        return f"识别结果: {input_data}"

def user_request(user_name, model, input_data):
    start = time.time()
    result=model.predict(input_data)

    end = time.time()
    cost=end-start
    record = {
        "user": user_name,
        "model":model.name,
        "cost": cost,
        "result": result
    }

    with lock:
        records.append(record)

a=TextModel("deepseek", "NLP")
b=ImageModel("seedance", "CV")

threads=[]
for i in range(1,3):
    t=threading.Thread(target=user_request,args=(f"User{i}",a,"你好，世界"))
    threads.append(t)

for i in range(3,5):
    t=threading.Thread(target=user_request,args=(f"User{i}",b,"一张🐱的图片"))
    threads.append(t)

for t in threads:
    t.start()

for t in threads:
    t.join()

for i in records:
    print(i)
```
